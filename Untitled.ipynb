```python
import mwparserfromhell
import re

def clean_value(value):
    # Use mwparserfromhell to parse the value
    wikicode = mwparserfromhell.parse(value)
    text = wikicode.strip_code()  # Convert to plain text

    # Remove Markdown-like syntax for bold/italic
    text = re.sub(r"'''''(.*?)'''''", r"\1", text)  # Bold and italic
    text = re.sub(r"'''(.*?)'''", r"\1", text)      # Bold
    text = re.sub(r"''(.*?)''", r"\1", text)        # Italic

    # Normalize whitespaces
    text = re.sub(r"\s+", " ", text).strip()

    return text
```

```python
import re
from bs4 import BeautifulSoup

def clean_value(value):
    # Remove HTML tags using BeautifulSoup
    value = BeautifulSoup(value, "html.parser").get_text()
    
    # Remove Markdown-like syntax for bold/italic
    value = re.sub(r"'''''(.*?)'''''", r"\1", value)  # Bold and italic
    value = re.sub(r"'''(.*?)'''", r"\1", value)      # Bold
    value = re.sub(r"''(.*?)''", r"\1", value)        # Italic

    # Remove links (keep the text)
    value = re.sub(r"\[\[(?:[^|\]]*\|)?([^\]]+)\]\]", r"\1", value)

    # Replace template syntax with a placeholder
    value = re.sub(r"\{\{[^\}]+\}\}", "", value)  # Remove templates

    # Normalize whitespaces
    value = re.sub(r"\s+", " ", value).strip()
    
    return value
```

```python
def preprocess_text(text):
    """Lowercase, remove punctuation, tokenize."""
    text = text.lower()
    tokens = word_tokenize(text)
    new_tokens = []
    for token in tokens:
        cleaned_token = token.translate(str.maketrans('', '', string.punctuation))
        if cleaned_token != '':
            new_tokens.append(cleaned_token)
    return new_tokens 
```

```python
def correct_sentence_spelling(tokens: List[str]) -> List[str]:
    spell = SpellChecker()
    misspelled = spell.unknown(tokens)
    for i, token in enumerate(tokens):
        if token in misspelled:
            corrected = spell.correction(token)
            if corrected is not None:
                tokens[i] = corrected
    return tokens
```

```python
def get_wordnet_pos(tag_parameter):

    tag = tag_parameter[0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    
    return tag_dict.get(tag, wordnet.NOUN)
```

```python
def lemmatization(tagged_doc_text):
    """Lemmatization."""
    lemmatized_doc_text = []
    for word, tag in tagged_doc_text:
        lemmatized_word = lemmatizer.lemmatize(word, pos=get_wordnet_pos(tag))
        lemmatized_doc_text.append(lemmatized_word)
    return lemmatized_doc_text 
```

```python
def search_engine(query):
    """Processes a query and returns the top matching documents."""
    preprocessed_query = preprocess_text(query)
    print(preprocessed_query)
    remove_stopwords_query = remove_stopwords(preprocessed_query)
    print(remove_stopwords_query)
    correct_query_spelling = correct_sentence_spelling(remove_stopwords_query)
    print(correct_query_spelling)
    pos_tag_query = pos_tag(correct_query_spelling)
    print(pos_tag_query)
    lemmatization_query = lemmatization(pos_tag_query)
    print(lemmatization_query)
    text_query = ' '.join(lemmatization_query)
    print(text_query)
    query_vector = vectorizer.transform([text_query]) 
    print(query_vector)
    cosine_similarities = cosine_similarity(tfidf_matrix, query_vector)
    print(cosine_similarities)
    
    cosine_similarities = (tfidf_matrix @ query_vector.T).toarray()
    top_indices = cosine_similarities.argsort()[0][-10:][::-1] 
    
    # return [documents[i] for i in top_indices]
    results_df = pd.DataFrame({
        "Document": [documents[i] for i in top_indices],
        "Similarity Score": cosine_similarities[top_indices].flatten() 
    })
    
    # return results_df

    results_data = []
    for i, doc_index in enumerate(top_indices):
        similarity = cosine_similarities[0, doc_index]
        results_data.append([i + 1, similarity, documents[doc_index]])

    table_headers = ["Rank", "Similarity", "Document"]
    print("Top 5 Results:")
    print(tabulate(results_data, headers=table_headers, tablefmt="grid"))  # 'grid' for a nice grid-like format
```

```python
query = "sturdy"
results = search_engine(query)
print("Top 5 Results:")
for result in results:
    print(result)
    print(results.to_string(index=False)) 
```
